```python
def experiment(num=2000):
    cars = []
    meters = []
    for n in range(num):
        places = np.zeros(100)
        centers = np.zeros(99)
        choices = np.random.randint(0, 98, size=1000)
        for i in choices:
            if (not places[i]) and (not places[i + 1]):
                places[i] = places[i + 1] = 1
                centers[i] = 1
        cars.append(places.sum() / 2)
        meters.append(100 - places.sum())
        
    return np.array(cars), np.array(meters)
```

```python
cars, meters = experiment()
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
def experiment2(p, num=500):
    cars = []
    meters = []
    honored = []
    dishonored = []
    for n in range(num):
        h = 0
        dh = 0
        drivers = np.random.choice([0, 1], p=[p, 1-p], size=1000)
        places = np.zeros(100)
        
        razm_places = np.zeros(50)
        
        centers = np.zeros(99)
        choices = np.random.randint(0, 98, size=1000)
        
        for i in choices:
            if drivers[i] == 0:
                if (not places[i]) and (not places[i + 1]):
                    dh += 1
                    places[i] = places[i + 1] = 1
                    centers[i] = 1
                    razm_places[i // 2] = 1
                    razm_places[(i + 1) // 2] = 1
            else:
                if (razm_places == 0).sum():
                    i = np.random.choice(np.arange(50)[razm_places == 0]) * 2
                    if (not places[i]) and (not places[i + 1]):
                        h += 1
                        places[i] = places[i + 1] = 1
                        centers[i] = 1
                        razm_places[i // 2] = 1
                        razm_places[(i + 1) // 2] = 1
                
        cars.append(places.sum() / 2)
        meters.append(100 - places.sum())
        honored.append(h)
        dishonored.append(dh)
        
        
    return np.array(honored), np.array(dishonored)
```

```python
percents = []
ps = np.linspace(0, 1, 21)
for p in ps:
    print(p)
    h, dh = experiment2(p)
    percents.append((h / (h + dh)) * 100)
    
```

```python
plt.plot(1 - ps, np.array(percents).mean(axis=1))
plt.xlabel('Вероятность приезда честного водителя')
plt.ylabel('процент честных водителей')
```
