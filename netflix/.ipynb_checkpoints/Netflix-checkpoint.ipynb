```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_shows = pd.read_csv('Netflix_Shows.csv', delimiter=',', encoding='CP1251')
```

```python
df_shows
```

```python
df_titles = pd.read_csv('netflix_titles.csv', delimiter=',')
```

```python
df_titles
```

```python
df_titles.index = df_titles.title
df_titles = df_titles.drop(['title'], axis=1)
df_titles
```

```python
df = df_shows.join(df_titles.drop(['rating'], axis=1), on='title')
```

```python
df
```

```python
sns.distplot(df.release_year)
```

```python
sns.boxplot(df.release_year, color='yellow')
```

```python
df = df[(df.show_id.notna()) & \
        (df.release_year >= 2015) & \
        (df['user rating score'].notna()) & \
        (df['country'].notna())].drop(\
                                      ['user rating size', 'title', 'director', 'cast', 'date_added', \
                                          'description', 'description', 'show_id',\
                                         'ratingDescription', 'ratingLevel', 'release year']
                                                                          , axis=1)
df
```

```python
df.duration = df.duration.apply(lambda x: int(x.split(' ')[0]))
```

```python
df
```

```python
plt.pie(df.groupby('type').count().iloc[:, 0].values, labels=df.type.unique(), colors=['red', 'black'])
```

```python
fig, ax = plt.subplots(figsize=(30,30))
b = sns.barplot(y="country", x="user rating score", data=df)
b.set_xlabel("user rating",fontsize=30)
b.set_ylabel("country",fontsize=20)
b.tick_params(labelsize=45)
```

```python
plt.plot(df[df.type == 'Movie'].groupby('duration')['user rating score'].mean())
plt.xlabel('duration')
plt.ylabel('user rating score')
```

```python
plt.plot(df[df.type == 'TV Show'].groupby('duration')['user rating score'].mean())
plt.xlabel('duration')
plt.ylabel('user rating score')
```

```python
b = sns.barplot(y="rating", x="user rating score", data=df)
b.set_xlabel("user rating")
b.set_ylabel("rating")
```

```python
categorical = ['rating', 'duration', 'release_year']
data = df.drop(categorical, axis=1)
for c in categorical:
    one_hot = pd.get_dummies(df[c], prefix=c) 
    data = data.join(one_hot)
```

```python
data
```

```python
data.country.values[:30]
```

```python
countries = []
for c in data.country.values:
    arr = c.split(", ")
    for a in arr:
        if a not in countries:
            countries.append(a)
countries
```

```python
for c in countries:
    data[c] = 0
```

```python
data
```

```python
cs = data.country.values
for i, c in enumerate(cs):
    for c1 in countries:
        if c1 in c:
            r = data.iloc[i]
            r[c1] = 1
            data.iloc[i] = r
```

```python
cats = []
for c in data['listed_in'].values:
    arr = c.split(", ")
    for a in arr:
        if a not in cats:
            cats.append(a)
cats
```

```python
for c in cats:
    data[c] = 0
```

```python
cs = data['listed_in'].values
for i, c in enumerate(cs):
    for c1 in cats:
        if c1 in c:
            r = data.iloc[i]
            r[c1] = 1
            data.iloc[i] = r
```

```python
data_movie = data[data.type == 'Movie'].drop('type', axis=1)
data_show = data[data.type == 'TV Show'].drop('type', axis=1)
```

```python
fig, ax = plt.subplots(figsize=(30,30))
corr = data_movie.corr().dropna(how='all', axis=1).dropna(how='all', axis=0)
sns.heatmap(corr, square=True)
```

```python
corr = corr['user rating score']
ind = np.argsort(corr.values)[::-1][1:10]
corr[ind]
```

```python
fig, ax = plt.subplots(figsize=(30,30))  
corr = data_show.corr().dropna(how='all', axis=1).dropna(how='all', axis=0)
sns.heatmap(corr, square=True)
```

```python
corr = corr['user rating score']
ind = np.argsort(corr.values)[::-1][1:15]
corr[ind]
```
