```python
import numpy as np
import cv2 # opencv
from matplotlib import pyplot as plt
import os
from PIL import Image, ImageDraw
import PIL
from skimage.io import imread
from skimage.filters import threshold_otsu
from skimage import measure
from skimage.measure import regionprops, regionprops_table
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
path = 'TAIWAN/'
# Получаем список файлов-фотографий
images = os.listdir(path)
```

```python
# Функция для поиска номеров
def find(img):
    # Библиотека для работы с графиками, чтобы выводить изображения
    import matplotlib.pyplot as plt
    # Считывает нашу картинку
    car_image = imread(path + img, as_gray=True)
    img = cv2.imread(path + img, 0)
    
    img = cv2.resize(img, (640, 480))
    # Выводим её размер
    gray_car_image = img
    # Создаём графики, на которых будут отображены наша фотография и её пороговое разделение
    # У нас будет таблица размера 1х2 с графиками (1 строка, 2 колонки)
    fig, (ax1, ax2) = plt.subplots(1, 2)
    # Привязываем наше изображение к левоме графику
    ax1.imshow(gray_car_image, cmap="gray")
    # Флаг, сообщающий нашли ли мы подходящие области или нет
    flag = 0
    
    plate_objects_cordinates = []
    xs = {}
    plate_like_objects = []
    areas = []
    borders = []
    # Перебираем пороговые изображения
    for binary_car_image in [
    gray_car_image > threshold_otsu(gray_car_image),
    *[cv2.adaptiveThreshold(img, 255 ,t_type_1, t_type_2, neighbour, 2) \
        for t_type_1 in [cv2.ADAPTIVE_THRESH_MEAN_C, cv2.ADAPTIVE_THRESH_GAUSSIAN_C]\
            for t_type_2 in [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV]\
                for neighbour in [5, 15, 11, 13]]
                            ]:
        
        # Привязываем бинарную картинку, к правому графику
        ax2.imshow(binary_car_image, cmap="gray")
        
        #показываем наши графики
        plt.show()

        label_image = measure.label(binary_car_image)
        min_height, max_height, min_width, max_width = (30, 80, 100, 150)
        
        min_area, max_area = min_width * min_height, max_width * max_height
        
        contours = measure.find_contours(label_image, 0.8)

        # Display the image and plot all contours found
        _, ax = plt.subplots()
        ax.imshow(label_image, cmap=plt.cm.gray)

        for n, contour in enumerate(contours):
            ax.plot(contour[:, 1], contour[:, 0], linewidth=1)
        # regionprops создает список свойств всех выделенных регионов в label_image (окантованных)
        for region in regionprops(label_image):
            
            # Координаты рамки
            min_row, min_col, max_row, max_col = region.bbox

            
            
            # Высота и ширина рамки
            region_height = max_row - min_row
            region_width = max_col - min_col

            if region_height * region_width < min_area or region_height * region_width > max_area:
                # Если у нас слишком маленькая/большая площадь для номера - не рассматриваем
                continue
            
            # Проверяем размер рамки, а также немного сужаем область, где может быть номерной знак
            if region_height >= min_height and region_height <= max_height and region_width >= min_width \
            and region_width <= max_width\
            and region_width > region_height and min_row > 25 and  min_col > 25\
                and  max_row < 400 and  max_col < 600:
                flag = 1
                # Добавляем подходящие элементы в список
                plate_like_objects.append(binary_car_image[min_row:max_row,
                                          min_col:max_col])
                # Добавляем координаты элементов в список
                plate_objects_cordinates.append((min_row, min_col,
                                                 max_row, max_col))
                if min_row in xs:
                    xs[min_row][0] += 1
                else:
                    xs[min_row] = [0, len(plate_like_objects) - 1]
                
                areas.append((max_row - min_row) * (max_col - min_col))
                
                # Создаём Bounding Box вокруг нашей области
                rectBorder = patches.Rectangle((min_col, min_row), max_col - min_col, max_row - min_row, edgecolor="red",
                                               linewidth=2, fill=False)
                borders.append(rectBorder)
                # Накладываем его на нашу картинку
                
        
            
            
    _, (ax3, ax3_1, ax4) = plt.subplots(1, 3)
    ax3.imshow(gray_car_image, cmap="gray")
    ax3_1.imshow(gray_car_image, cmap="gray")
    
    if(flag == 1):
        areas = np.array(areas)
        print(areas.max())
        for ind in range(len(plate_like_objects)):
            ax3.add_patch(borders[ind])
        ind = 0
        #ind = areas.argmax()
        rect = plate_objects_cordinates[ind]
        min_row, min_col, max_row, max_col = rect
        r = patches.Rectangle((min_col, min_row), max_col - min_col, max_row - min_row, edgecolor="red",
                                               linewidth=2, fill=False)
        ax3_1.add_patch(r)
        ax4.imshow(plate_like_objects[ind], cmap='gray')
        # Показываем картинку, если нашли подходящую область
        plt.show()
        return





[find(img) for img in images]
```
